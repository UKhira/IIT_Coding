```python
# import the CSV file
import pandas as pd

data_frame = pd.read_csv('/content/5DATA002W.2 Coursework Dataset(25012025v6.0).csv');

data_frame
```

```python
# View Dimensions
data_frame.shape
```

```python
# Get an idea of data
data_frame.info()
```

```python
# To identify unusual values in data_frame
data_frame.describe()
```

```python
# Looking at the results where Age < 0 and Age > 200 and Tumor_Size < 0  values are highly unlikable
# remove those outliers
# data_frame = data_frame[(data_frame['Age'] >= 0) & (data_frame['Age'] <= 200) & (data_frame['Tumor_Size'] > 0)]
data_frame.drop(data_frame[(data_frame['Age'] < 0) | (data_frame['Age'] > 200) | (data_frame['Tumor_Size'] < 0)].index, inplace = True)
data_frame.describe()
```

```python
# Remove unwanted records to build model

data_frame.drop(['Patient_ID' ,'Month_of_Birth', 'Occupation'], axis = 1, inplace = True)
data_frame
```

```python
# Identify null records
data_frame.isnull().sum()
```

```python
# Remove null records from data_frame (The maximum null count recorded is 9 which is very low compare to tolal records)

data_frame.dropna(inplace=True)
data_frame.isnull().sum()
```

```python
data_frame['Mortality_Status'] = data_frame['Mortality_Status'].str.upper()
data_frame['Mortality_Status'].unique()
```

```python
# Changing categorical variable
data_frame['Sex'] = data_frame['Sex'].map({'Female': 2, '1':1})
data_frame['T_Stage'] = data_frame['T_Stage'].map({'T1': 1, 'T2': 2, 'T3': 3, 'T4': 4})
data_frame['N_Stage'] = data_frame['N_Stage'].map({'N1': 1, 'N2': 2, 'N3': 3})
data_frame['6th_Stage'] = data_frame['6th_Stage'].map({'IIA': 1, 'IIB': 2, 'IIIA': 3, 'IIIB': 4, 'IIIC': 5})
data_frame['Differentiated'] = data_frame['Differentiated'].map({'Poorly differentiated': 1, 'Moderately differentiated' : 2, 'Well differentiated' : 3, 'Undifferentiated' : 4})
data_frame['A_Stage'] = data_frame['A_Stage'].map({'Regional': 1, 'Distant': 2})
data_frame['Estrogen_Status'] = data_frame['Estrogen_Status'].map({'Positive': 1, 'Negative' : 0})
data_frame['Progesterone_Status'] = data_frame['Progesterone_Status'].map({'Positive': 1, 'Negative' : 0})
data_frame['Mortality_Status'] = data_frame['Mortality_Status'].map({'ALIVE': 1, 'DEAD': 0})

data_frame.head()
```
