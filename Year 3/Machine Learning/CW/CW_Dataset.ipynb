```python
# import the CSV file
import pandas as pd

dataset = pd.read_csv('/content/5DATA002W.2 Coursework Dataset(25012025v6.0) (2).csv');

dataset.head()
```

```python
# View Dimensions
dataset.shape
```

```python
# Get an idea of data
dataset.info()
```

```python
# Identify null records
dataset.isnull().sum()
```

```python
# Validate Data Columns (Only applicables)
# Months cannot exceed 12

month_set = dataset[(dataset.Month_of_Birth > 12)]
month_set.head()
```

```python
# Remove unwanted records

dataset.drop('Occupation', axis = 1, inplace = True)
dataset.drop('Patient_ID', axis = 1, inplace = True)
dataset.drop('Month_of_Birth', axis = 1, inplace = True)
dataset.head()
```

```python
# Remove null records from dataset

dataset.dropna(inplace=True)
dataset.isnull().sum()
```

```python
dataset['Mortality_Status'] = dataset['Mortality_Status'].str.upper()
dataset['Mortality_Status'].unique()
```

```python
# Changing categorical variable
dataset['Sex'] = dataset['Sex'].map({'Female': 2, '1':1})
dataset['T_Stage'] = dataset['T_Stage'].map({'T1': 1, 'T2': 2, 'T3': 3, 'T4': 4})
dataset['N_Stage'] = dataset['N_Stage'].map({'N1': 1, 'N2': 2, 'N3': 3})
dataset['6th_Stage'] = dataset['6th_Stage'].map({'IIA': 1, 'IIB': 2, 'IIIA': 3, 'IIIB': 4, 'IIIC': 5})
dataset['Differentiated'] = dataset['Differentiated'].map({'Poorly differentiated': 1, 'Moderately differentiated' : 2, 'Well differentiated' : 3, 'Undifferentiated' : 4})
dataset['A_Stage'] = dataset['A_Stage'].map({'Regional': 1, 'Distant': 2})
dataset['Estrogen_Status'] = dataset['Estrogen_Status'].map({'Positive': 1, 'Negative' : 0})
dataset['Progesterone_Status'] = dataset['Progesterone_Status'].map({'Positive': 1, 'Negative' : 0})
dataset['Mortality_Status'] = dataset['Mortality_Status'].map({'ALIVE': 1, 'Dead': 0})

dataset.head()
```

```python
# Split records to train and test data
dataset.shape
```

```python
#This will save the dataset without the raws indeces
dataset.to_csv(r'/content/Prepared_Data_Set.csv', index=False)
```
